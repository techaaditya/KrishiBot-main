```python
import pandas as pd
import numpy as np
from constants import CROPS_USED
import numpy as np
from numpy import array
import pandas as pd
class Model:
    _df = pd.read_csv("./data/data.csv")

    crops_used=CROPS_USED

    df_filtered=_df[_df['label'].isin(crops_used)]

    summary_dict = df_filtered.groupby("label").mean().T.to_dict()

    d = {k: np.array(list(v.values())) for k,v in summary_dict.items()}

    @classmethod
    def report(cls):
        """Prints the report on the model vs dataset."""
        from sklearn.metrics import classification_report
        test = {k: array(list(v.values())) for k,v in cls.df_filtered.drop("label",axis=1).T.to_dict().items()}
        predictions = array([Model.topn(v, 1)[0][0] for v in test.values()])
        actual = array(cls.df_filtered.label)
        print(classification_report(actual, predictions))
        
    @classmethod
    def predict(cls,params: np.ndarray):
        """
        Order of params: ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
        """
        return { k: np.linalg.norm(v-params) for k,v in cls.d.items()}

    @classmethod
    def topn(cls,params, n: int):
        "Returns top_keys, sorted_prediction_dict"
        sorted_prediction_dict = dict(sorted(Model.predict(params).items(), key=lambda item: item[1]))
        return list(sorted_prediction_dict.keys())[:n],sorted_prediction_dict
        
```
